### 1. Away_Def_Pass.csv

```python
Away_Def_Pass = pd.read_csv("Away_Def_Pass.csv")
Away_Def_Pass = Away_Def_Pass.set_index('game_id')
#Away_Def_Pass
```

```python
pca_dimension_kept = PCA().fit(Away_Def_Pass)
plt.plot(np.cumsum(pca_dimension_kept.explained_variance_ratio_))
plt.xlabel('The Number of Components')
plt.ylabel('The Cumulative Explained Variance')
```

```python
vec_explnd_var_ratio = pca_dimension_kept.explained_variance_ratio_.cumsum()
itemindex = np.where(vec_explnd_var_ratio >= 0.99)
dimension_kept = itemindex[0][0]+1
print("The number of components which has 99% cumulative explained variance is:", dimension_kept)
```

```python
pca_dimension_redc = PCA(n_components=dimension_kept)
pca_dimension_redc.fit(Away_Def_Pass)
Away_Def_Pass_PCA = pd.DataFrame(pca_dimension_redc.transform(Away_Def_Pass))
Away_Def_Pass_PCA.columns=['Away_Def_Pass_PCA_'+ str(i) for i in range(1,dimension_kept+1)]
print("The size of this transformed dataset is:", Away_Def_Pass_PCA.shape)
print(Away_Def_Pass_PCA)
```

### Plot the ROC curve

```python
from sklearn.utils import shuffle
from sklearn.metrics import roc_curve, auc
```

```python
def ROC(label,result):
    # Compute ROC curve and area the curve
    Y = np.array(label)
    fpr, tpr, thresholds = roc_curve(Y, result)
    roc_auc = auc(fpr, tpr)
    print("Area under the ROC curve : %f" % roc_auc)

    # Plot ROC curve
    #pl.clf()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

```python
ROC(y_test,y_predict_logistic_reg)
```

### Obtain the precision recall curve and obtain the average precision

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
import matplotlib.pyplot as plt
```

```python
Average_Precision = average_precision_score(y_test, y_predict_logistic_reg)

Precision, Recall, _ = precision_recall_curve(y_test, y_predict_logistic_reg)

plt.step(Recall, Precision, color='b', alpha=0.2,
         where='post')
plt.fill_between(Recall, Precision, step='post', alpha=0.2,
                 color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve: Average Precision={0:0.2f}'.format(Average_Precision))
```

```python
print('The average precision is {}.'.format(Average_Precision))
```

### Find the optimal threshold to maximize the F1 score

```python
from sklearn.metrics import f1_score
```

```python
threshold = np.arange(0, 1, 0.1).tolist()
f_1 = []

for i in threshold:
    y_predict_logis_reg_th = np.where(classification_logistic_reg.predict_proba(X_test)[:,1] > i, 1, 0)
    f_1_score = f1_score(y_test, y_predict_logis_reg_th)
    f_1.append(f_1_score)
    
a = np.argmax(f_1)

threshold_selected = threshold[a]
f_1_max = f_1[a]

print('The F1 Score is maximized when the threshold is',threshold_selected,'.', 
      'And the corresponding F1 Score is', f_1_max, '.')
```

```python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import classification_report
from sklearn.metrics import matthews_corrcoef
```

```python
C_values = [0.01,0.1,1,10,100]
pipe_steps = [
    ('svc', SVC())
]
pipeline = Pipeline(pipe_steps)
cv_split = ShuffleSplit(n_splits=5, train_size=0.7, test_size=0.25)
param_grid = {
    #'svc__cache_size' : [0.01,0.1,1,1.5],
    'svc__C': C_values,
    'svc__kernel' : ['rbf', 'linear'],
    #'svc__gamma' : gamma_values,
    #'svc__degree' : [0.001,0.005,0.01],
}

grid = GridSearchCV(pipeline, param_grid, cv=cv_split)
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
